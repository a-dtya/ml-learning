**Gradient Descent**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
print(X)
```

```python
y = (X-5)**2 + 3
print(y)
```

```python
plt.style.use("seaborn")
plt.plot(X,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Here gradient  = derivative of y wrt x, i.e 2(x-5)
assuming learning rate = 0.1 and number of iterations to be 50

```python
lr=0.1
for i in range(50):
  grad=2*(X-5)
  X=X-lr*grad
  print(X)
plt.plot(X,y)  
plt.show()
```

```python
lr=0.1
for i in range(30):
  grad=2*(X-5)
  X=X-lr*grad
  y=(X-5)**2+3
  plt.scatter(X,y)
plt.show()  
```
